#Expansions of functions
Suppose you are asked to solve the wave equation in one dimension:
$$
\frac{1}{c^2} \frac{\partial^2}{\partial t^2} \psi(x,t) = \frac{\partial^2}{\partial x^2} \psi(x,t) .
$$
This equation determines for example the transverse motion of an elastic string, or the longitudinal air displacement in an organ pipe..
The solution to this problem depends on the boundary conditions imposed. Usually, the time $t$ assumes any value: $-\infty < t < \infty$. If we do neither impose a limit on the position $x$, then we have the solution
$$
\psi(x, t) = f(x-ct) + g(x+ct)
$$
for *any* smooth one-dimensional functions $f$ and $g$. The $f$-solution represents a wave running to the right, whereas $g$ is a left-running wave.

It is important to note that the wave equation is linear, hence any linear combination of solutions is also a solution. More precisely, if $\psi(x,t)$ and $\chi(x,t)$ both satisfy the wave equation,
then $\lambda \psi(x,t) + \mu \chi(x,t)$ is also a solution. 
Such linear combination may involve infinite numbers of solutions and the result may be a solution of a form which is very different from the above (with $f$ and $g$).

Now let us consider the wave equation on a finite interval $x \epsilon [0, L]$ with boundary condition
$\psi(0,t) = \psi(L,t) = 0 $ for *all* $t$. To find the solution we can use a trick called 'separation of variables'. This works as follows. Let us consider a solution of the form
$$
\psi(x,t) = \phi(x) \theta(t).
$$
Putting this into the wave equation we find 
$$
\phi(x) \frac{1}{c^2} \frac{d^2}{d t^2} \theta(t) = \theta(t)  \frac{d^2}{d x^2} \phi(x).
$$
Dividing the left and right hand side of this equation by $\psi(x,t) = \phi(x) \theta(t)$ we find
$$
\frac{1}{c^2}\frac{d^2 \theta/d t^2}{\theta} =  \frac{d^2 \phi/d x^2}{\phi}
$$
where we have left out the $x$ and $t$ dependencies in order to avoid cluttered notation.
In the last equation, the left hand side depends on $t$ only, whereas the right hand side exclusively depends on $x$. This can only be the case if both sides are a constant which depends neither on $x$ nor on $t$. We call this constant $\lambda$ to arrive at
$$
\frac{1}{c^2}\frac{d^2 \theta(t)}{d t^2} = \lambda \theta(t); \\
\frac{d^2 \phi(x)}{d x^2} = \lambda \phi(x). \\
$$
Let us now investigate the cases $\lambda=0$, $\lambda<0$ and $\lambda>0$.

1. For $\lambda = 0$, the solution for $\phi$ is $ax+b$. The zero-boundary conditions for $x=0$ and $x=L$ can only be satisfied for $a-b=0$, i.e. the string is standing still in its equilibrium state.
2. For $\lambda<0$ we find for $\phi$ the solution 
   $$
   \phi(x) = A \sin (kx) + B \cos(kx),
   $$
   with $k^2 = -\lambda$.
   which satisfies the boundary conditions only when $B=0$ and $kL = n\pi$, $n=1, 2, 3, \ldots$. 
3. For $\lambda>0$ the solution for $\phi$ is $$
   A e^{qx} + B e^{-qx} $$
   which does not satisfy the boundary conditions (verify this explicitly!).
   
We see that there is an infinite number of solutions $\sin(kx)$ where $k$ assumes the discrete set of values $kL=n\pi$ and $k^2 = -\lambda$. What about the time dependence? Since $\lambda$ is negative, we find the 
solution
$\theta(t) = C \sin(\omega t) + D \cos(\omega t)$ with $\omega^2 = -c^2 \lambda = c^2 k^2$. Without imposing any initial condition, all values for $C$ and $D$ are allowed. 

We have seen that the second derivatives of $\phi$ and $\theta$ are both equal to the constant $\lambda$. The boundary conditions $\psi(0,t) = \psi(L,t) = 0 $ restricts 
$\lambda$ to the values
$$
\lambda_n = - k_n^2 , 
$$
where $k_n = n \pi/L$, so
$$
\lambda_n = - \frac{n^2\pi^2}{L^2}.
$$
We may therefore label the prefactors $C$ and $D$ each by $n$, to obtain a solution
$$
\sin(n \pi x/L) \left[ C_n \sin(\omega_n t) + D_n \cos(\omega_n t)\right] 
$$ where $C_n$ and $D_n$ are arbitrary and 
$$
\omega^2_n = -c^2 \lambda_n = c^2 k_n^2.
$$
Any solution will be a (possibly infinite) sum of these solutions.

For a particular solution, the parameters $C_n$ and $D_n$ are in practice determined by the *initial conditions*. If we release the string from rest at $t=0$, we know that the velocity of all points on the string is $0$. This immediately implies that $C_n=0$ for all $n$. 
Using the linearity of the wave equation, we see that the most general solution for a string released *from rest* at $t=0$ is:
$$
\psi(x,t) = \sum_{n=1}^\infty D_n \sin(k_n x) \cos(\omega_n t).
$$
The actual value of the $D_n$ is determined by the particular shape of the string at $t=0$. 

Now let us find out how we can find these $D_n$. For this purpose, the language of linear algebra is very useful. 
Let us analyse the set of functions $\sin (k_n x)$ with $k_n = n \pi/L$. We consider these as basis functions in a 
Hilbert space. The inner product in that space is 
$$
\bra{\chi}\psi\rangle = \int_0^L \chi^*(x) \psi(x) dx.
$$
Since we will restrict ourselves to real functions, the star $*$ is not required. 
We can then normalize the basis functions - the normalized functions are
$$
\phi_n(x) = \sqrt{\frac{2}{L}} \sin(k_n x)
$$
where of course $n=1, 2, \ldots$. 
> **Exercise** Check that this is the correct normalization.

It can be shown that two such functions with different index $n$ are orthogonal so that we have
$$
\bra{\phi_n} \phi_m \rangle = \delta_{nm},
$$
where $\delta_{nm}$ is the usual Kronecker delta-function.

Now consider the initial shape of the string, which we call $f(x)$. This shape must satisfy the boundary conditions. We can expand this function in the basis functions (using $\tilde{D}_n$ because we have normalised the wave functions):
$$
f(x) = \bra{x} f\rangle = \sum_{n=1}^\infty \tilde{D}_n \bra{x} \phi_n \rangle
$$
or, simply
$$
\ket{f} = \sum_{n=1}^\infty \tilde{D}_n \ket{\phi_n}.
$$
The $\tilde{D}_m$ are found by taking the inner product with $\phi_m$ from the left:
$$
\bra{\phi_m} f \rangle = \tilde{D}_m.
$$
The terms with $n\neq m$ in the expansion on the right hand side have disappeared as a result of the delta function. 
Writing out the inner product as an integral, we have
$$
\tilde{D}_n = \sqrt{\frac{2}{L}} \int_0^L \sin(n\pi x/L) f(x) dx.
$$
What we have seen is quite miraculous: if we pluck a string in the middle, its shape is given by
$ f(x) = a (|L/2-x| + L/2).$ This can apparently also be written as an infinite sum over sine functions. This is demonstrated in the picture below, where the triangular shape of a plucked string just before it is released, 
is approximated by a sum of up to 11 cosines. 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1.0, 1.0, 200)
ft = np.ndarray((200, 11),dtype=float)
for n in range(11):
    k = (2*n+1)*0.5*np.pi
    prefac = 2.0/k**2
    ft[:,n] = prefac*np.cos(k*x)
    y = np.sum(ft[:,:n], axis = 1)
    plt.plot(x,y)
plt.show()
# y = np.sum(ft, axis = 1)
#plt.plot(x,y)
```

The information about the full solution is contained in the function $f$. However, the *same* information is
stored in the (infinite) sequence $\tilde{D}_n$. The $\tilde{D}_n$ are the *Fourier coefficients* and transforming from $f$ to and from $\tilde{D}_n$ is called *Fourier transformation*.

In the foregoing analysis we have emphasised the algebraic structure of the Fourier transformation procedure. Interestingly, we can view the equation
$$
\frac{1}{c^2} \frac{\partial^2 }{\partial x^2}\phi(x) = \lambda \phi(x)
$$
as an *eigenvalue problem* of the hermitian operator $d^2/dx^2$! The eigenvactors are the functions $\sin(n\pi x/L)$. The fact that the operator is hermitian immediately demonstrates that indeed these functions are orthogonal (recall that this orthogonality of the eigenfunctions is a property of the eigenvalues of a hermitian operator). 

The fact that we only have the $\tilde{D}_n$ in our expansion, and no coefficients deriving from the $C_n$, is 
due to the fact that we assumed that the string was released from rest. If we take the viewpont that the string 
has shape $f=0$ at $t=0$, but that the speed of the points of the string has shape $g(x)$ (i.e. $g(x)$ is the vertical speed of the string at $t=0$), we would obtain and expansion
$$
\psi(x,t) = \sum_{n=1}^\infty C_n \sin(k_n x) \sin(\omega_n t).
$$
The remainder of the analysis would then be the same, with coefficients $\omega_n C_n$ (or $\omega_n \tilde{C}_n$ for normalized 
basis functions) being determined by
$$
\omega \tilde{C}_n = \sqrt{\frac{2}{L}} \int_0^L \sin(n\pi x/L) g(x) dx.
$$

If the boundary conditions are periodic, both sine and cosine solutions are allowed in the spatial part. In that case, it turns out convenient to use $\phi_n(x) = \exp(\ii k_n x)/\sqrt{L}$ as basis functions, where now 
$$k_n = 2\pi n/L
$$ and $n$ is allowed to run over all integers. Let us calculate the inner product of these:
$$
\bra{\phi_n} \phi_m \rangle = \frac{1}{L} \int_0^L e^{\ii (k_n-k_m) x} dx. 
$$
For $n=m$, we immediately see that the result is $L$. For $n\neq m$, we can work out the real and imaginary parts separately, and we find the value $0$. 
> **Exercise** Verify these statements, i.e. show that $\bra{\phi_n} \phi_m \rangle = \delta_{nm}$. 

Therefore, an orthomormal basis set on the periodic interval with length $L$ is 
$$
\bra{x} \phi_n\rangle = \frac{1}{\sqrt{L}} e^{\ii k_n x}.
$$
The function $f$ can be expanded in terms of the exponentials:
$$
f(x) = \sum_{n=-\infty}^\infty \tilde{f}_n e^{\ii k_n x}.
$$
To find the Fourier coefficients $\tilde{f}_n$, we take the inner product from the left with a particular basis function $\exp(\ii k_m x)$ (which then becomes $\exp(-\ii k_m x)$ inside the integral because the inner product involves the complex conjugate of the function on the left):
$$
\tilde{f}_m = \frac{1}{L} \int_0^L e^{-\ii k_m x} f(x) dx.
$$

Note
-----
For the periodic Fourier transform, the choice of the boundaries is not critical: we could also have the $x$ run from $-L/2$ to $L/2$ for example. In that case, the Fourier sum would not change, and the coefficients would be calculated as 
$$
\tilde{f}_m = \frac{1}{L} \int_{-L/2}^{L/2} e^{-\ii k_m x} f(x) dx.
$$

#Fourier transforms on the real axis
Now we consider the same problem on the real axis $-\infty < x < \infty$. 
This proceeds in a similar way as for the finite interval; however, the boundary condition is replaced by the condition that the solution does not go to $\pm \infty$ for the entire axis. 
This restricts the spatial part of the solution again to $\sin(kx)$ and $\cos(kx)$. It turns out convenient to use complex exponentials for this case, even if the position of the string is real everywhere. 

The solutions are of the form:
$$
\exp[\ii (kx - \omega t)].
$$

The functions $\exp(\ii k x)$ can still be used as basis functions. However, these basis functions cannot be labelled by a discrete index $n$ (or $k_n$) - this is replaced by the *continuous index* $k$. So we now denote 
the $\ket{\phi_n}$ of the previous section by $\ket{k}$. 
The problem is now to calculate 
$$
\bra{k} k'\rangle = \int \exp[\ii (k - k') x] dx.
$$
Let us calculate this first on a finite but large interval where we can use the expressions of the previous section. There we saw that for $k\neq k'$, the result vanishes. Only for $k=k'$ we had a nonzero result and we then
arrive at $L$. But we want to increase $L$ to infinity! 

To find out how we can arrive at a sensible Fourier transform, we start from a function $f(x)$ defined on the periodic interval $[-L/2, L/2]$. The Fourier transform is now a discrete transform and we have
$$
\tilde{f}_n = \frac{1}{L} \int_{-L/2}^{L/2} f(x) e^{-\ii kx} dx.
$$
In this case, $k_n=2\pi n/L$, so the spacing between adjacent $k_n$-values is $\Delta k = 2\pi /L$. 
We see that this Fourier transform gives $\tilde{f}_0 = \frac{1}{L} \int f(x) dx$, i.e. the average of the function $f$ on the interval $[0,L]$. 

We can rewrite the pre-factor before the integral in terms of the spacing $\Delta k$:
$$
\tilde{f}_n = \frac{\Delta k}{2\pi} \int_{-L/2}^{L/2} f(x) e^{-\ii k_n x} dx.
$$
Now consider the inverse transformation:
$$
f(x) = \sum_n \tilde{f}_n e^{\ii k_n x} = \sum_n \frac{\Delta k}{2\pi} \left( \int_{0}^{L} f(x') e^{-\ii k_n x'} dx' \right) e^{\ii k_n x}.
$$
The expression $\sum_n \; \Delta k \; e^{-\ii k_n x'} e^{\ii k_n x} $
is a good approximation for
$$
\int_{-\infty}^\infty e^{\ii k (x-x')} dk;
$$
the smaller $\Delta k$ (hence, the larger $L$), the better this approximation will be. So we conclude that for $L\rightarrow \infty$
$$
f(x) = \frac{1}{2\pi} \int_{-\infty}^\infty dk\; e^{\ii k x}Â \int_{-\infty}^{\infty} f(x') e^{-\ii k x'} dx'.
$$
We now define the Fourier transform $\tilde{f}(k)$ of the function $f(x)$ as 
$$
\tilde{f}(k) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^\infty f(x) e^{-\ii k x} dx,
$$
so that 
$$
f(x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^\infty \tilde{f}(k) e^{\ii k x} dk.
$$
The last two equations define the Fourier transform and its inverse respectivelly.

Notes
------

1. Applying the Fourier transform and its inverse: $$f(x) = \frac{1}{2\pi} \int_{-\infty}^\infty dk \; e^{\ii k x} \int_{-\infty}^\infty f(x') e^{-\ii k x'} dx'k$$ can be reformulated, after defining the function $$D(x) = \frac{1}{2\pi} \int_{-\infty}^\infty \; e^{\ii k x}\; dk,$$ as $$f(x) = \frac{1}{2\pi} \int_{-\infty}^\infty D(x-x') f(x') dx'.$$ The function $D(x)$ is called the *Dirac delta function*. It is used extremely often in physics and engineering. It can (crudely) be visualised as a function which is zero everywhere except at $x=0$, where it assumes the value $\infty$.
2. The definition of the Fourier transform is not unique. In fact, often it is defined as $$\tilde{f}(k) = \int_{-\infty}^\infty f(x) e^{-\ii k x} dx,$$ i.e. with a different prefactor. The inverse is then $$ f(x) = \frac{1}{2\pi} \int_{-\infty}^\infty e^{\ii k x} \tilde{f}(k) dx.$$ The product of the prefactors before the integrals of the transform and its inverse should thus alway be $1/(2\pi)$.
3. Similarly, the Fourier transform and its inverse may be defined with different signs in the integrals, hence $$\tilde{f}(k) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^\infty f(x) e^{\ii k x} dx,$$ and $$ f(x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^\infty e^{- \ii k x} \tilde{f}(k) dx.$$



#Linear algebra and Fourier transform
The Fourier transform of a function$f$ is $\tilde{f}$. A relevant question we may ask is how the norms of
$f$ and $\tilde{f}$ are related:
$$
\bra{\tilde{f}}\tilde{f}\rangle = \int_{-\infty}^\infty \left|\tilde{f}(k)\right|^2 dk = 
\int_{-\infty}^\infty dk \frac{1}{2\pi} \int_{-\infty}^\infty e^{-\ii k x} f(x) dx \int_{-\infty}^\infty e^{-\ii k x'} f(x') dx' .
$$
Now we use the fact that
$$
\frac{1}{2\pi} \int_{-\infty}^\infty e^{\ii k (x-x')} dk = \delta (x-x')
$$
to rewrite the triple integral in the rightmost term as
$$
\int_{-\infty}^\infty \left| f(x) \right|^2 dx = \bra{f} f \rangle.
$$
This leads to conclusion that the norm of the Fourier transform of a function equals that of the original function. 

A formulation in terms of inner products of the Fourier transform can be arrived at by the identification
$$
\bra k x \rangle = \frac{1}{\sqrt{2\pi}} e^{\ii kx}.
$$
We furthermore set $$
\bra x f \rangle = f(x) $$ and $$
\bra k f \rangle = \tilde{f}(k) .$$
Then, the Fourier transform can be written in the form
$$
f(x) = \bra{x} f \rangle = \int_{-\infty}^\infty \bra{x} k \rangle \bra{k} f \rangle dk,
$$
whereas the inverse reads:
$$
\tilde{f}(k) = \bra{k} f \rangle = \int_{-\infty}^\infty \bra{k} x \rangle \bra{x} f \rangle dx.
$$
Also, we see that the unit operator has matrix elements
$$ \bra{x} 1 \ket{x'} = \int_{-\infty}^\infty \bra{x} k \rangle \bra{k} x' \rangle dk = 
\frac{1}{2\pi} \int_{-\infty}^\infty e^{\ii k(x-x')} dk = \delta(x-x').
$$

An interesting property concerns the Fourier transform of the derivative of a function:
$$
\int_{-\infty}^\infty f'(x) e^{-\ii kx}dx = f(x) e^{\ii k x}\left. \right|^\infty_{-\infty} + \ii k \int_{-\infty}^\infty f(x) e^{\ii kx}dx.
$$
Assuming that $f(\infty) = f(- \infty) = 0$, we see that the Fourier transform of the derivative of a function can 
be written as $ \ii k$ times the Fourier transform of the original function.

#Multi-dimensional Fourier transforms
The continuous Fourier transform can easily be generalized to higher dimensions. For exampl the 3D transform reads
$$
\tilde{f}(\bfk) = \frac{1}{(2\pi)^{3/2}} \int f(\bfr) e^{-\ii \bfk \cdot \bfr} d^3r
$$
with the inverse
$$
f(\bfr) = \frac{1}{(2\pi)^{3/2}} \int f(\bfk) e^{\ii \bfk \cdot \bfr} d^3k.
$$





#Expansions in other functions

A Fourier transform is an expansion of a function in terms of basis functions of the form $\exp(\ii kx) /\sqrt{2\pi}$. These basis functions are eigenfunctions of the Hermitian operator $d^2/dx^2$. Expansions in other functions also occur, where the basis functions are also related to eigenvalue problems. A nice example is given by the wave equation for a circular drum. The wave equation has the form 
$$
\frac{1}{c^2} \frac{\partial^2}{\partial t^2} \psi(\bfr,t) = \nabla^2 \psi(\bfr,t),
$$
where $\bfr$ is a two-dimensional position vector and $\nabla^2$ is the Laplace operator in two dimensions.
Performing the usual separation of the variables $\bfr$ and $t$, we end up with an eigenvalue equation for the spatial wave function:
$$
\nabla^2 \chi(\bfr) = \lambda \chi(\bfr).
$$
The drumhead is clamped at its circular boundary. We call $a$ the radius of the drumhead, that is, $\chi(|\bfr|=a) = 0$. It is convenient to use polar coordinates $r$ and $\varphi$. 
The Laplace operator $\nabla^2$ reads in polar coordinates:
$$
\nabla^2 \chi(r,\varphi) = \frac{1}{r} \frac{\partial}{\partial r} \left(r\frac{\partial}{\partial r}\right) \chi(r,\varphi) + \frac{1}{r^2} \frac{\partial^2}{\partial \varphi^2} \chi(r,\varphi).
$$
Now we perform separation of the variables $r$ and $\varphi$:
$$
\chi (r, \varphi) = R(r) \Phi(\varphi).
$$
We put this form into the eigenvalue equation $\nabla^2 \chi = \lambda \chi$ and divide the left and right hand side by $\chi$ to obtain:
$$
\frac{1}{R} \frac{1}{r} \frac{\partial}{\partial r} \left(r\frac{\partial}{\partial r} R(r) \right) + 
\frac{1}{\Phi} \frac{1}{r^2} \frac{\partial^2}{\partial \varphi^2} \Phi(r, \varphi) = \lambda.
$$

This equation has the form of a function depending on $r$ plus a function depending on $r$ and $\varphi$ equal a 
constant ($\lambda$). In view of the fact that there is only one term involving $\varphi$ as a variable, we conclude that that dependence should disappear, and we therefore have 
$$\frac{1}{\Phi} \frac{1}{r^2} \frac{d^2}{d r^2} \Phi(\varphi) = \frac{\mu}{r^2}.
$$
This equation can easily be solved - we find 
$$\Phi(\varphi) = e^{\pm \ii \sqrt\mu \varphi}. 
$$
Note that because of the geometry of the problem, $\Phi$ must be periodic in $\varphi$ with period $2\pi$. 
Therefore, we have $\sqrt \mu = m$, $m=0, \pm 1, \pm 2, \ldots$. 
We are then left with 
$$
\frac{1}{R} \frac{1}{r} \frac{d}{d r} \left(r\frac{d}{d r} R(r) \right) - \frac{m^2}{r^2} = \lambda.
$$
Multiplying this by $r^2 R$ and working out the derivatives leads to
$$
r^2 R'' + r R' - m^2 R = \lambda r^2 R
$$
where the prime indicates a derivative with respect to $r$. Note that $\lambda$ may be negative 
(and is expected to be negative by analogy to the wave equation in one dimension). 
By introducing the new variable $x = \sqrt{-\lambda} r$, this is cast into the form
$$
x^2 R'' + x R' +(x^2- m^2) R = 0,
$$
where the prime now indicates a derivative with respect to $x$. This equation is known as the *Bessel equation*. 
For future reference, we set $k=\sqrt{-\lambda}$.

Its solutions which behave regularly at the origin are the Bessel functions. These functions  are not expressed simply in terms of exponentials, sines or cosines. The Bessel functions are called $J_m(x)$. A few of them are shown in the figure below.


```python

import scipy.special as sp

x = np.linspace(0.0, 20.0, 400)
j0 = sp.jv(0,x)
j1 = sp.jv(1,x)
j2 = sp.jv(2,x)
j3 = sp.jv(3,x)

j0l = plt.plot(x, j0, label='$J_0$')
j1l = plt.plot(x, j1, label='$J_1$')
j2l = plt.plot(x, j2, label='$J_2$')
j3l = plt.plot(x, j3, label='$J_3$')
plt.legend(loc='upper right')
plt.axhline(0, color='black')
```

Obviously, for the Bessel functions that are a correct solution to our problem, we must have $J_m(x) = 0$ for
$x = ka$. Now recall that for the time dependent part of the solution, we have a sine or a cosine of $\omega t$, with $\omega = ck$. This follows from the fact that this time dependence, $\Theta(t)$, satisfies
$$ \frac{d^2 \Theta}{dt^2} = \lambda.$$
So, the zeroes of the Bessel functions tell us which frequencies are excited when we strike a drum!

You will always find that Bessel functions occur in the context of cylindrically symmetric problems. An example is 
the electric potential of a wire in 3D. Along the direction of the wire, the potential is constant, but perpendicular to it, it takes the form of a Bessel function (or a superposition of Bessel functions).

If we play the same game in 3D, we would find *spherical Bessel functions* for the radial coordinate, multiplied by the so-called *spherical Harmonics* $Y_l^m(\vartheta, \varphi)$. The latter are combinations of sine and cosines of the angles $\vartheta$ and $\varphi$.
